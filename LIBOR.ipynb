```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotlibor(dates, proj):
    nmonths = len(dates)
    plt.figure(figsize = (10, 5), dpi = 100)
    plt.plot(range(nmonths), proj, color = 'blue', label = 'Projection')
    plt.xlabel('Months from ' + dates[0], fontsize = 18)
    plt.ylabel('LIBOR', fontsize = 18)
    plt.grid()
    plt.legend()
    plt.tick_params(labelsize = 14)
    plt.show()
    return
```

```python
libbloom = pd.read_excel('libor8-10.xlsx').values
```

```python
dates, proj = libbloom[1:].T[0], libbloom[1:].T[2]
```

```python
plotlibor(dates, proj)
```
