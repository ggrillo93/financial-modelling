```python
def viewportcf(names, alldates, amort, propcf, interest, loanpay, debt):
    """ Returns a dataframe object with the cashflow per period for the individual properties in a portfolio, as well
        as the total cashflow per period.
    Inputs: names = list/array containing the identifying name for each of the properties
            allcf = individual cashflow per period for each of the properties
            alldates = all dates at which portfolio cashflows take place """

    rows = np.array([np.arange(len(debt)), alldates.date])
    totpropcf = np.sum(propcf, axis = 0)
    totcf = totpropcf + amort + interest + loanpay
    Df = debt + amort + loanpay
    Df[0] = debt[1]
    moneys = np.vstack([debt, amort, propcf, interest, loanpay, Df, totcf])
    rows = np.vstack([rows, convtomoney(moneys)])
    rowlabels = ['Month #', 'Dates', 'Starting debt', 'Amortization'] + names + ['Interest payment', 'Loan cashflow from buy/sell', 'Final debt', 'Total cash flow']
    df = pd.DataFrame(rows, index = rowlabels, columns = ['']*len(debt))
    # df.columns.name = 'Month #'
    return df

def convtomoney(arr):
    """ Converts array of floats/integers into an array of formatted money objects. """
    floatarr = np.array(arr)
    nelem = len(floatarr.flatten())
    dim = floatarr.shape
    monarr = np.zeros(nelem, dtype = object)
    for i in range(nelem):
        monarr[i] = format_money(Money(amount = str(floatarr.flatten()[i]), currency = 'USD'), locale = 'en_US')
    return monarr.reshape(dim)
```

```python
names = ['Bicentenario', 'Santa Laura']
alldates = pd.date_range(start = np.datetime64('2018-08-31'), periods = 10, freq = 'M')
amort = np.random.randn(10)
propcf = np.random.randn(2, 10)
interest = np.random.randn(10)
loanpay = np.random.randn(10)
debt = np.random.randn(10)
viewportcf(names, alldates, amort, propcf, interest, loanpay, debt)
```
